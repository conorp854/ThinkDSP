```python
import thinkdsp
import thinkplot
```

```python
cos_sig = thinkdsp.CosSignal(freq=440, amp=1.0, offset=0)
sin_sig = thinkdsp.SinSignal(freq=880, amp=0.5, offset=0)
```

```python
mix = sin_sig + cos_sig
```

```python
wave  = mix.make_wave(duration=0.5, start=0,framerate=11025)
```

```python
wave

```

```python
wave.plot()
```

```python
mix.plot()
```

```python
period = mix.period
segment = wave.segment(start=0, duration=period*4)
segment.plot()
```
